```python
# GaussianNB Algorithm for 20 different values and using varr_smothing parameter (TASK 5-C)
from sklearn.naive_bayes import GaussianNB                            
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split         # The useful libraries and datasets already imported above  but i want to show them seperately
from sklearn.linear_model import LinearRegression

iris = sns.load_dataset('iris')
X_iris = iris.drop('species', axis=1)
y_iris = iris['species']

Xtrain, Xtest, ytrain, ytest = train_test_split(X_iris, y_iris,random_state=1)
                                               

predicted_values=[1,2,25,45,76,333,45,67,111111,40,60,32,34,35,9,0,6,87,0.76,0.89877]
aimed_accuracy=0                                                       
for i in range(20):                                                  
  X=predicted_values[i]          
  sample = GaussianNB(var_smoothing=X)                                
  sample.fit(Xtrain, ytrain)                                          
  y_sample = sample.predict(Xtest)                                                          
  accuracy=(accuracy_score(ytest, y_sample))                        

  a=pd.DataFrame([accuracy]);
  b=a.max()

print('The Max Value:',b);
```

```python
# Comparing of GaussianNB algorithm with other two algorithms (Task 5-D)
from sklearn.naive_bayes import MultinomialNB   # importing MultinominalNB for comparing results of GaussianNB

import warnings
warnings.filterwarnings("ignore")

iris = sns.load_dataset('iris')
X_iris = iris.drop('species', axis=1)
y_iris = iris['species']

Xtrain, Xtest, ytrain, ytest = train_test_split(X_iris, y_iris, random_state=1)

predicted_values=[1,2,25,45,76,333,45,67,111111,40,60,32,34,35,9,0,6,87,0.76,0.89877]
aimed_accuracy=0                                                       
for i in range(20):                                                  
  X=predicted_values[i]          
  sample = MultinomialNB(alpha=X)                # i changed GaussianNB to MultinominalNB for style of using i changed varr_smothing to alpha                
  sample.fit(Xtrain, ytrain)                                          
  y_sample = sample.predict(Xtest)                                                          
  accuracy=(accuracy_score(ytest, y_sample))         ;               

  a=pd.DataFrame([accuracy]);
  b=a.max() ;

print('The Max Value:',b);                      # The new max value
```

```python
from sklearn.naive_bayes import BernoulliNB    # importing BernoulliNB for comparing results of GaussianNB

import warnings
warnings.filterwarnings("ignore")

iris = sns.load_dataset('iris')
X_iris = iris.drop('species', axis=1)
y_iris = iris['species']

Xtrain, Xtest, ytrain, ytest = train_test_split(X_iris, y_iris, random_state=1)

predicted_values=[1,2,25,45,76,333,45,67,111111,40,60,32,34,35,9,0,6,87,0.76,0.89877]
aimed_accuracy=0                                                       
for i in range(20):                                                  
  X=predicted_values[i]          
  sample = BernoulliNB(fit_prior =X)                # i changed GaussianNB to BernoulliNB for style of using i changed varr_smothing to fit_prior                
  sample.fit(Xtrain, ytrain)                                          
  y_sample = sample.predict(Xtest)                                                          
  accuracy=(accuracy_score(ytest, y_sample))         ;               

  a=pd.DataFrame([accuracy]);
  b=a.max() ;

print('The Max Value:',b);                      # The new max value
```
